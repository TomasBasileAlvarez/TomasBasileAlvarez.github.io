## Dibujar Conujnto de Mandelbrot

```python
def matriz_compleja(xmin, xmax, ymin, ymax, densidad):
    'Crear el plano complejo en cierto rango'
    reales = np.linspace(xmin, xmax,  densidad)
    imags = np.linspace(ymin, ymax,  densidad)
    return reales[np.newaxis, :] + imags[:, np.newaxis] * 1j

def estable(c, iters):
    #empezamos desde z=0
    z = 0
    
    #Aplicamos la funcion z_{n+1} = z_n**2 + c iters veces
    for i in range(iters):
        z = z ** 2 + c
    
    #Si el valor absoluto no es mayor a 2, no diverge
    return abs(z) <= 2

def miembros(c, iters):
    c_mask = estable(c, iters)
    return c[c_mask]


def dibujar_mandelbrot(xi,xf,yi,yf,densidad=1000,iters=40):
    c = matriz_compleja(xi, xf , yi , yf, densidad)
    
    img = plt.imshow(estable(c, iters),
           cmap="binary",
           origin='lower',vmin=0,vmax=1,extent=[xi,xf,yi,yf]) 
    
    plt.gca().set_aspect("equal")
    plt.title('Conjunto de Mandelbrot')
    # plt.axis("off")
    plt.tight_layout()
    plt.show()

dibujar_mandelbrot(-2,1,-1,1)
```

```python
def logistic(x, r,a):
    'Mapeo Logistico'
    return r * x * (1 - x)

def gauss(xn,b,a=7.5):
    'Mapeo de Gauss'
    xn1 = b + np.exp(-a*xn**2)
    return(xn1)


def periodo(b,a=7.5,x0=0,itera=200,impr=False,func=gauss,deep=15):
    'Funcion para calcular el periodo de un mapeo'
    lista=[x0]
    for i in range(itera):
        x = func(lista[-1],b,a)
        if np.abs(x)>10 or math.isnan(np.real(x)) == True:
            if impr == True:
                print(lista)
            return(0)
        lista.append(x)
    if impr == True:
        print(lista)  
        
    lista = np.array(lista[-deep-2:])
    lista = lista-lista[-1]
    for i in range(2,deep+1):
        if np.abs(lista[-i])<0.005:
            if np.abs(lista[-i-1]-lista[-2])<0.005:
                    return(i-1)
    return(13)

def diagrama(b,a=7.5,x0=0,itera=200,func=gauss,impr=False):

    lista = [x0]

    for i in range(itera):
        lista.append(func(lista[-1],b))
    
    x= [ele.real for ele in lista]
    y = [ele.imag  for ele in lista]

    plt.plot(x,y,'-.r*')
    plt.show()
    plt.plot(x[-itera//10:],y[-itera//10:],'-r*')
    plt.show()
    print(periodo(b,a=a,x0=x0,itera=itera,func=func,impr=impr))

    
def dibujar(a=7.5,x0=0,itera=50,func=gauss,xi=-1,xf=1,
            yi=-1,yf=1,N=150,nombre='im',deep=15):
    
    pru=np.zeros((N,N),complex)
    mat = np.zeros((N,N))
    j = 0
    for x1 in np.linspace(xi,xf,N):
        i=0
        for x2 in np.linspace(yi,yf,N):
            pru[i][j] = x1+x2*1j
            per = periodo(x1+x2*1j,a=a,x0=x0,itera=itera,deep=deep,func=func)
        
            mat[i][j] = per
            i+=1        
        
        j+=1
    if nombre=='im':
        img = plt.imshow(mat,origin='lower',vmin=0,vmax=deep,extent=[xi,xf,yi,yf]) 
        plt.colorbar(img)
        plt.show()
    else:
        img = plt.imshow(mat,origin='lower',vmin=0,vmax=deep,extent=[xi,xf,yi,yf]) 
        plt.savefig(nombre + '-' + str(np.random.randint(999))+'.png',bbox_inches='tight')
        plt.colorbar(img)
        plt.show()
    
def bif_real(xi=0,xf=4,x0 = 1e-5,a=0,iterations=1000,last=100,n=10000,func=logistic,
            titulo = 'Diagrama de bifurcación del mapa de Gauss con a= ',
            xlim=0,ylim=0,
        ylab= 'Valor final', xlab='Valor de b'):
    'Funcion para dibujar el diagrama de bifurcacion'
    titulo = titulo+ str(a)
                         
    if xlim == 0:
        xlim = [xi,xf]
    

    r = np.linspace(xi, xf, n)
    x = x0 * np.ones(n)
#     fig, ax = plt.subplots()
    res = []
    for i in range(iterations):
    
        x = func(x, r,a)

        if i >= (iterations - last):
            plt.plot(r, x, ',k', alpha=.25)
            
    

    plt.title(titulo)
    plt.xlim(xlim)
    
    plt.xlabel(xlab)
    plt.ylabel(ylab)

    if ylim!=0:
        plt.ylim(ylim)
        
    
    
    plt.show()
#     ax.show()  

def dibujar_telarana(r, x0, n, func=logistic, ax=None):
    'Dibujar el diagrama de telaraa'
    # Dibujamos la función y la línea y=x
    t = np.linspace(0, 1)
    ax.plot(t, logistic(t,r,0), 'k', lw=2.5)
    ax.plot([0, 1], [0, 1], 'k', lw=2.5)

    #Aplicar y=f(x) varias veces, empezando desde x=x0
    x = x0
    for i in range(n):
        y = func(x,r,0)
        # Dibujar las lineas
        ax.plot([x, x], [x, y], 'k', lw=1)
        ax.plot([x, y], [y, y], 'k', lw=1)
        ax.plot([x], [y], 'ok', ms=10,
                alpha=(i + 1) / n)
        
        #Actualizar el valor de x
        x = y
    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 
                1)
    ax.set_title("Mapa Logístico: " + f"$r={r:.1f}, \, x_0={x0:.1f}$")


```

### Dibujar  diagrama de bifurcación para parámetro complejo

```python
import matplotlib.cm as cm    
def dibujar(a=7.5,x0=0,itera=50,func=gauss,xi=-1,xf=1,
            yi=-1,yf=1,N=150,nombre='im',deep=15,mapa=cm.Greys,
           titulo=''):
    
    
    pru=np.zeros((N,N),complex)
    mat = np.zeros((N,N))
    j = 0
    for x1 in np.linspace(xi,xf,N):
        i=0
        for x2 in np.linspace(yi,yf,N):
            pru[i][j] = x1+x2*1j
            per = periodo(x1+x2*1j,a=a,x0=x0,itera=itera,deep=deep,func=func)
        
            mat[i][j] = per
            i+=1        
        
        j+=1
        
    if nombre=='im':
        img = plt.imshow(mat,origin='lower',vmin=0,vmax=deep,extent=[xi,xf,yi,yf],cmap=mapa) 
        plt.colorbar(img)
        plt.title(titulo)
        plt.xlabel('Parte real de b')
        plt.ylabel('Parte compleja de b')
        plt.show()
    else:
        img = plt.imshow(mat,origin='lower',vmin=0,vmax=deep,extent=[xi,xf,yi,yf],cmap=mapa) 
        plt.savefig(nombre + '-' + str(np.random.randint(999))+'.png',bbox_inches='tight')
        plt.colorbar(img)
        plt.title(titulo)
        plt.xlabel('Parte real de b')
        plt.ylabel('Parte compleja de b')

        plt.show()
```
