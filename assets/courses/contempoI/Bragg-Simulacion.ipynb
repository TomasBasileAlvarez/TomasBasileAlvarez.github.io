```python
import numpy as np
import matplotlib.pyplot as plt

# Pondremos el origen de coordenadas (0,0) en la posición en la que se
# encuentra el emisor


emisor = np.array([0,0,0])
# Definimos el valor de la distancia entre átomos a:
a = 4
    
# La distancia del emisor al cubo "De" y del detector al cubo "Dd"
# El ángulo que hace el emisor respecto al cubo "thi" y del receptor "thf"
De = 50
Dd = 50

#Definimos el ángulo del cubo respecto al eje x, que define al plano
# 0 para plano (1,0,0), 45 para (1,1,0) y 26.6 para (1,2,0)
ang = 26.6
    
# Definimos la longitud de onda "lamb", el número de onda k y frec. angular "w"
lamb = 2.85    
k= 2*np.pi/lamb
c = 3e8
w = c/lamb*2*np.pi

#Vamos a iterar sobre el ángulo th que hace tanto el emisor como receptor
#respecto a los planos.
#En "ángulos" guardaremos los ángulos y en "resultados" los resultados 
#para cada ángulo
angulos= []
resultados = []


for th in range(10,81):
    angulos.append(th)

    #Creamos las posiciones de los átomos. #Empezamos calculando la del centro:
    centro = np.array([De*np.sin(th*np.pi/180),De*np.cos(th*np.pi/180),0])

    #Definimos los vectores de traslación
    vec1 = np.array([a*np.cos(ang*np.pi/180),a*np.sin(ang*np.pi/180),0] )
    vec2 = np.array([-a*np.sin(ang*np.pi/180),a*np.cos(ang*np.pi/180),0] )
    vec3 = np.array([0,0,a])

    #Calculamos la posición del detector
    detector = centro+np.array([-Dd*np.sin(th*np.pi/180),Dd*np.cos(th*np.pi/180),0])

    #La posición de la esquina inferior izquierda del cubo
    esquina = centro-2*vec1 - 2*vec2 -2*vec3
    
    #Vamos a guardar las posiciones de los átomos
    atomos = []

    #Vamos guardando las posiciones de cada átomo
    for n in range(5):
        for m in range(5):
            for l in range(5):
                atomos.append(esquina + n*vec1 + m*vec2 + l*vec3)
            
    #Vamos a dibujar las cosas, pero solo para un ángulo, el de 30 grados
    if th == 30:
        atomosx = []
        atomosy = []
        for at in atomos:
            atomosx.append(at[0])
            atomosy.append(at[1])    
        plt.scatter(atomosx,atomosy)
        plt.plot(emisor,'ko')
        plt.plot(detector,'ro')
        plt.axis('equal')
        plt.title('Vista desde arriba del arreglo para ángulo theta de 30 grados')

        plt.xlim([-5,100])
        plt.ylim([-5,100])
        plt.xlabel('Posición en eje x [cm]')
        plt.ylabel('Posición en eje y [cm]')
        plt.show()
    
            
    # Vamos calculando el campo eléctrico en el detector y lo guardamos en 
    #tot, que empieza en 0. Para calcularlo, lo promediaremos en el tiempo.
    tot = 0
    
    for t in np.arange(0,2*np.pi*2/w,np.pi*2/w/100):
        # a la aportación al tiempo t le llamamos res
        res = 0
        
        # Calculamos la aportación de cada átomo
        for vec in atomos:
            mag = np.linalg.norm(vec)
            mag2 = np.linalg.norm(detector-vec)
        
            #Sumamos la aportación de este átomo
            res+= 1 / (mag * mag2) * np.exp(1j*(k*mag + k*mag2-w*t))
        
        #Al total le sumamos el campo cuadrado de cada átomo / 100
        #para promediarlo sobre los 100 tiempos
        
        tot += np.abs(res**2)/100
    
    resultados.append(tot)
    
plt.plot(angulos,resultados)   
plt.xlabel(r'Ángulo $\theta$')
plt.ylabel('Intensidad')
plt.title(r'Intensidad vs ángulo para ángulo $\alpha$ = ' + str(ang))
```
